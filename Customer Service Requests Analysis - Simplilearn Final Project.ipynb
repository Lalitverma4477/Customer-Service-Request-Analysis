# Data exploratory analysis:

```python
# bar plot to show the number of null values in each column of the DataFrame

plt.figure(figsize=(10, 6))
null_counts = data.isnull().sum()
null_counts.plot(kind="bar")
plt.title('Number of Null Values in Each Column')
plt.xlabel('columns')
plt.ylabel('Frequency')
plt.show()
```

```python
# Remove the records whose Closed Date values are null
data.dropna(subset=["Closed Date"], inplace= True)
```

```python
data.shape
```

```python
data.info()
```

```python
#check datatype
data['Closed Date'].dtypes
```

```python
# Calculate the time elapsed in closed and creation date

data['Closed Date'] = pd.to_datetime(data["Closed Date"])
data['Created Date'] = pd.to_datetime(data["Created Date"])
data['time_elapsed'] = data['Closed Date'] - data['Created Date']


data['time_elapsed']
```

```python
data['Closed Date'].dtypes
```

```python
#Convert the calculated date to seconds to get a better representation
data['time_elapsed'] = (data['Closed Date'] - data['Created Date']).dt.total_seconds()
data['time_elapsed']
```

```python
# frequency plot for the complaints in each city

city_counts = data['City'].value_counts()

plt.figure(figsize=(10, 6))
city_counts.plot(kind='bar')
plt.title('Complaints in Each City')
plt.xlabel('City')
plt.ylabel('Frequency')
plt.show()
```

```python
# scatter plot of the concentration of complaints across Brooklyn

brooklyn_df = data[data['City'] == 'BROOKLYN']

plt.figure(figsize=(10, 6))
sns.scatterplot(data=brooklyn_df, x='Longitude', y='Latitude')
plt.title('Concentration of Complaints across Brooklyn')
plt.show()
```

```python
# hexbin plot of the concentration of complaints across Brooklyn

brooklyn_df = data[data['City'] == 'BROOKLYN']

plt.figure(figsize=(10, 6))
plt.hexbin(brooklyn_df['Longitude'], brooklyn_df['Latitude'] , gridsize=30, cmap='Blues')
plt.title('Concentration of Complaints across Brooklyn')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar(label='Count')
plt.show()
```

```python
#bar graph to show the types of complaints

complaints_count = data['Complaint Type'].value_counts()

plt.figure(figsize=(11,6))
complaints_count.plot(kind ="bar")
plt.xticks(rotation=45, ha='right')
plt.title('Types of Compaints')
plt.xlabel('complaints Types')
plt.ylabel('frequeny')

for i, count in enumerate(complaints_count):
    plt.text(i, count, str(count), ha='center', va='bottom')
    
plt.show()
```

```python
# Display the various types of complaints in each city

complaints_by_city = data.groupby(['City', 'Complaint Type']).size().reset_index(name='Count')


complaints_by_city
```

```python
# Creating a DataFrame, df_new, which contains cities as columns and complaint types in rows

df_new = pd.pivot_table(data, index='Complaint Type', columns='City', aggfunc='size', fill_value=0)

df_new
```

```python
# Drawing a chart that shows the types of complaints in  each city in a single chart, where different colors show the  different types of complaints

complaints_by_city = data.groupby(['City', 'Complaint Type']).size().unstack()

complaints_by_city.plot(kind='bar', stacked=True, figsize=(15, 10))

plt.title('Types of Complaints by City')
plt.xlabel('City')
plt.ylabel('Count')

plt.legend(title='Complaint Type')

plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```

```python
# Sort the complaint types based on the average Request_Closing_Time grouping them for different locations

avg_request_closing_time = data.groupby(['City', 'Complaint Type'])['time_elapsed'].mean()

sorted_complaint_types = avg_request_closing_time.groupby('City').apply(lambda x: x.sort_values(ascending = True))
sorted_complaint_types
```

```python
# Visualize the average of Request_Closing_Time

average_request_closing_time = data.groupby('Complaint Type')['time_elapsed'].mean()

plt.figure(figsize=(9, 10))
average_request_closing_time.plot(kind='pie')

plt.title('Average Request Closing Time by Complaint Type')
plt.xlabel('Complaint Type')
plt.ylabel('Average Request Closing Time (in hours)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
